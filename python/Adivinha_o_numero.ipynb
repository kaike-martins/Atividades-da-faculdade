```python
# @title Atividade
import random

recordistas = []
jogar_novamente = "s"

while jogar_novamente == "s":
    n = random.randrange(0, 21)
    acertou = False
    tentativas = 0

    while not acertou and tentativas < 15:
        palpite = int(input("Digite o número: "))
        tentativas += 1
        if palpite < n:
            print("O número é maior")
        elif palpite > n:
            print("O número é menor")
        else:
            acertou = True
            print(f"Parabéns! Você acertou em {tentativas} tentativas!")

    if not acertou:
        print(f"Você perdeu! O número era {n}")
    else:
        nome = input("Qual seu nome? ").title()
        recordistas.append({"nome": nome, "tentativas": tentativas})
        recordistas.sort(key=lambda x: x["tentativas"])
        recordistas = recordistas[:3]
        print("Placar atual:")
        for i, r in enumerate(recordistas):
            print(f"{i+1}. {r['nome']} - {r['tentativas']} tentativas")

    jogar_novamente = input("Deseja jogar novamente? (s/n): ").lower()

print("Jogo encerrado! Placar final:")
for i, r in enumerate(recordistas):
    print(f"{i+1}. {r['nome']} - {r['tentativas']} tentativas")
```
