```python
print("Bem vindo(a) quitanda que Anda")
print("No momento vendemos somente quatro variáveis de frutas")

# -------- MORANGO --------
mor = int(input("Quantos kilos de morango você comprará? "))
while mor < 0:
    print("Quantidade inválida!")
    mor = int(input("Digite novamente: "))

if mor <= 5:
    pm = 5
elif mor > 5 and mor < 10:
    pm = 3.5
else:
    pm = 2

totmor = mor * pm
print("Você vai gastar de morango:", totmor, "reais")

# -------- ABACAXI --------
aba = int(input("Quantos kilos de abacaxi você comprará? "))
while aba < 0:
    print("Quantidade inválida!")
    aba = int(input("Digite novamente: "))

if aba <= 5:
    pa = 10
elif aba > 5 and aba < 10:
    pa = 7.5
else:
    pa = 6.8

tota = aba * pa
print("Você vai gastar de abacaxi:", tota, "reais")

# -------- PERA --------
pera = int(input("Quantos kilos de pera você comprará? "))
while pera < 0:
    print("Quantidade inválida!")
    pera = int(input("Digite novamente: "))

if pera <= 5:
    pp = 8.5
elif pera > 5 and pera < 10:
    pp = 7
else:
    pp = 5.5

totp = pera * pp
print("Você vai gastar de pera:", totp, "reais")

# -------- MAÇÃ --------
maca = int(input("Quantos kilos de maça você comprará? "))
while maca < 0:
    print("Quantidade inválida!")
    maca = int(input("Digite novamente: "))

if maca <= 5:
    pma = 6.5
elif maca > 5 and maca < 10:
    pma = 5
else:
    pma = 3.75

totma = maca * pma
print("Você vai gastar de maçã:", totma, "reais")

# -------- TOTAL --------
compra = totmor + tota + totp + totma
compraf = compra

# -------- PAGAMENTO --------
print("\nForma de pagamento:")
print("1 - Dinheiro (10% desconto)")
print("2 - Débito (5% desconto)")
print("3 - Crédito (2x sem juros)")

pag = int(input("Escolha: "))

if pag == 1:
    compraf = compraf * 0.9
elif pag == 2:
    compraf = compraf * 0.95
elif pag == 3:
    pass  # sem desconto

# -------- ENTREGA --------
print("\nForma de entrega:")
print("1 - Entrega (10,00R$)")
print("2 - Retirar no local")

entrega = int(input("Escolha: "))

if entrega == 1:
    compraf = compraf + 10

# -------- RESULTADO FINAL --------
print("\nTotal final da compra:", compraf, "reais")

# -------- PARCELAMENTO --------
if pag == 3:
    parcela = compraf / 2
    print("Pagamento em 2x de", parcela, "reais")
```
